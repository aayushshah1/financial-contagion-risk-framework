# MCA Records Analysis

This notebook analyzes MCA records from `data_consolidation/data/mca_records.json`, focusing on:
- Total records
- Total companies by listing status (Listed/Unlisted)
- Total companies by company status (e.g., Active, Strike Off)


## 1. Import Required Libraries
Import pandas and json libraries for data manipulation and analysis.

```python
# Import Required Libraries
import pandas as pd
import json
```

## 2. Load MCA Records JSON Data
Load the `mca_records.json` file into a pandas DataFrame, handling any necessary JSON normalization.

```python
# Load MCA Records JSON Data
with open('../data_consolidation/data/mca_records.json', 'r', encoding='utf-8') as f:
    records = [json.loads(line) for line in f]

df = pd.json_normalize(records)
df.head()
```

## 3. Calculate Total Records
Compute and display the total number of records in the dataset.

```python
# Calculate Total Records
total_records = len(df)
print(f"Total records: {total_records}")
```

## 4. Calculate Total Companies by Listing Status
Group and count companies by the 'Listingstatus' field (e.g., Listed, Unlisted).

```python
# Calculate Total Companies by Listing Status
listing_counts = df['Listingstatus'].value_counts(dropna=False)
print("Total Companies by Listing Status:")
print(listing_counts)
```

## 5. Calculate Total Companies by Company Status
Group and count companies by the 'CompanyStatus' field (e.g., Active, Strike Off).

```python
# Calculate Total Companies by Company Status
status_counts = df['CompanyStatus'].value_counts(dropna=False)
print("Total Companies by Company Status:")
print(status_counts)
```
